```python
emissions_capita.drop(emissions_capita.index[266:], axis=0, inplace=True)
emissions_capita.drop('Series Code', axis=1, inplace=True)
emissions_capita.drop('Country Code', axis=1, inplace=True)
emissions_capita.drop('Series Name', axis=1, inplace=True)
emissions_capita.drop('2020 [YR2020]', axis=1, inplace=True)
emissions_capita.drop('2021 [YR2021]', axis=1, inplace=True)
emissions_capita2 = emissions_capita.replace('..', np.nan)

emissions_capita = emissions_capita2.astype(
    {'1991 [YR1991]': 'float',
     '1990 [YR1990]': 'float',
     '1992 [YR1992]': 'float',
     '1993 [YR1993]': 'float',
     '1994 [YR1994]': 'float',
     '1995 [YR1995]': 'float',
     '1996 [YR1996]': 'float',
     '1997 [YR1997]': 'float',
     '1998 [YR1998]': 'float',
     '1999 [YR1999]': 'float',
     '2000 [YR2000]': 'float',
     '2001 [YR2001]': 'float',
     '2002 [YR2002]': 'float',
     '2003 [YR2003]': 'float',
     '2004 [YR2004]': 'float',
     '2005 [YR2005]': 'float',
     '2006 [YR2006]': 'float',
     '2007 [YR2007]': 'float',
     '2008 [YR2008]': 'float',
     '2009 [YR2009]': 'float',
     '2010 [YR2010]': 'float',
     '2011 [YR2011]': 'float',
     '2012 [YR2012]': 'float',
     '2013 [YR2013]': 'float',
     '2014 [YR2014]': 'float',
     '2015 [YR2015]': 'float',
     '2016 [YR2016]': 'float',
     '2017 [YR2017]': 'float',
     '2018 [YR2018]': 'float',
     '2019 [YR2019]': 'float'})
emissions_capita.columns = ['Country Name',1990,1991,1992,1993,1994,1995,1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
```

```python
years = list(range(1990,2020))
```

```python
emissions_capita[emissions_capita['Country Name']=='Poland']
```

```python
Polska_emisja = emissions_capita[emissions_capita['Country Name']=='Poland'].iloc[:,1:].transpose()
Polska_emisja.columns = ['emissions']
```

```python
Niemcy_emisja = emissions_capita[emissions_capita['Country Name']=='Germany'].iloc[:,1:].transpose()
Niemcy_emisja.columns = ['emissions']
```

```python
Francja_emisja = emissions_capita[emissions_capita['Country Name']=='France'].iloc[:,1:].transpose()
Francja_emisja.columns = ['emissions']
```

```python
Finlandia_emisja = emissions_capita[emissions_capita['Country Name']=='Finland'].iloc[:,1:].transpose()
Finlandia_emisja.columns = ['emissions']
```

```python
Dania_emisja = emissions_capita[emissions_capita['Country Name']=='Denmark'].iloc[:,1:].transpose()
Dania_emisja.columns = ['emissions']
```

```python
Estonia_emisja = emissions_capita[emissions_capita['Country Name']=='Estonia'].iloc[:,1:].transpose()
Estonia_emisja.columns = ['emissions']
```

```python
Chiny_emisja = emissions_capita[emissions_capita['Country Name']=='China'].iloc[:,1:].transpose()
Chiny_emisja.columns = ['emissions']
```

```python
fig, ax1 = plt.subplots()
ax1.stackplot(years,Polska_emisja.emissions, Niemcy_emisja.emissions, Francja_emisja.emissions, Finlandia_emisja.emissions, Dania_emisja.emissions, Estonia_emisja.emissions, labels=['Polska', 'Niemcy', 'Francja', 'Finlandia', 'Dania', 'Estonia'])
plt.legend()
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax1.plot(years, Polska_emisja.emissions, label='Poland')
ax1.plot(years, Niemcy_emisja.emissions, label='Germany')
ax1.plot(years, Francja_emisja.emissions, label='France')
ax1.plot(years, Finlandia_emisja.emissions, label = 'Finland')
ax1.plot(years, Estonia_emisja.emissions, label = 'Estonia')
ax1.plot(years, Dania_emisja.emissions, label = 'Denmark')
ax1.plot(years, Chiny_emisja.emissions, label = 'China')
ax1.set_ylabel('CO2 emissions (metric tons per capita)')
ax1.set_xlabel('Year')
ax1.set_title('EUROPE + CHINA')
plt.legend()
plt.show()
```
