```python
# Data reader 
def get_image_batch(num_images=16, img_size=(224, 224)):

    # 8 losowo wybranych gatunków
    species = np.random.choice(species_list, 8, replace=False)
    
    # Dla każdego z 8 losowo wybranych gratunków losujemy po 16 zdjęć
    image_files = []
    for s in species:
        species_dir = os.path.join(train_dir, s)
        species_files = os.listdir(species_dir)
        species_files = [os.path.join(species_dir, f) for f in species_files]
        species_files = np.random.choice(species_files, num_images, replace=False)
        image_files.extend(species_files)
    
    # Wgrywanie zdjęć
    images = []
    for file in image_files:
        image = cv2.imread(file)
        image = cv2.resize(image, img_size)
        images.append(image)
    
    # Zdjęcie do np.array
    images = np.array(images)
    
    # Normalizacja
    images = images.astype('float32') / 255.0
    
    return images
```
