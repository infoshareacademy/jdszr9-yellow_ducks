```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
emissions = pd.read_csv('emissions.csv')
```

```python
pd.set_option('display.max_rows', 500)
```

```python
emissions.head()
```

```python
emissions.columns
```

```python
emissions.drop(emissions.index[266:], axis=0, inplace=True)
emissions.drop('Series Code', axis=1, inplace=True)
emissions.drop('Country Code', axis=1, inplace=True)
emissions.drop('Series Name', axis=1, inplace=True)
emissions.drop('2020 [YR2020]', axis=1, inplace=True)
emissions.drop('2021 [YR2021]', axis=1, inplace=True)
emissions2 = emissions.replace('..', np.nan)

emissions = emissions2.astype(
    {'1991 [YR1991]': 'float',
     '1990 [YR1990]': 'float',
     '1992 [YR1992]': 'float',
     '1993 [YR1993]': 'float',
     '1994 [YR1994]': 'float',
     '1995 [YR1995]': 'float',
     '1996 [YR1996]': 'float',
     '1997 [YR1997]': 'float',
     '1998 [YR1998]': 'float',
     '1999 [YR1999]': 'float',
     '2000 [YR2000]': 'float',
     '2001 [YR2001]': 'float',
     '2002 [YR2002]': 'float',
     '2003 [YR2003]': 'float',
     '2004 [YR2004]': 'float',
     '2005 [YR2005]': 'float',
     '2006 [YR2006]': 'float',
     '2007 [YR2007]': 'float',
     '2008 [YR2008]': 'float',
     '2009 [YR2009]': 'float',
     '2010 [YR2010]': 'float',
     '2011 [YR2011]': 'float',
     '2012 [YR2012]': 'float',
     '2013 [YR2013]': 'float',
     '2014 [YR2014]': 'float',
     '2015 [YR2015]': 'float',
     '2016 [YR2016]': 'float',
     '2017 [YR2017]': 'float',
     '2018 [YR2018]': 'float',
     '2019 [YR2019]': 'float'})
emissions.columns = ['Country Name',1990,1991,1992,1993,1994,1995,1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
```

```python
emissions.sort_values(2019, ascending = False)
```

```python
years = list(range(1990,2020))
```

```python
Swiat_emisja = emissions[emissions['Country Name']=='World'].iloc[:,1:].transpose()
Swiat_emisja.columns = ['emissions']
```

```python
Chiny_emisja = emissions[emissions['Country Name']=='China'].iloc[:,1:].transpose()
Chiny_emisja.columns = ['emissions']
```

```python
USA_emisja = emissions[emissions['Country Name']=='United States'].iloc[:,1:].transpose()
USA_emisja.columns = ['emissions']
```

```python
Indie_emisja = emissions[emissions['Country Name']=='India'].iloc[:,1:].transpose()
Indie_emisja.columns = ['emissions']
```

```python
EU_emisja = emissions[emissions['Country Name']=='European Union'].iloc[:,1:].transpose()
EU_emisja.columns = ['emissions']
```

```python
fig, ax1 = plt.subplots()
ax1.plot(years, Swiat_emisja.emissions)
ax1.set_ylabel('CO2 emissions (kt)')
ax1.set_xlabel('Year')
ax1.set_title('WORLD')
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax1.plot(years, EU_emisja.emissions, label='EU')
ax1.plot(years, Chiny_emisja.emissions, label='China')
ax1.plot(years, Indie_emisja.emissions, label='India')
ax1.plot(years, USA_emisja.emissions, label='USA')
ax1.set_ylabel('CO2 emissions (kt)')
ax1.set_xlabel('Year')
ax1.set_title('EU vs China')
plt.legend()
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax1.stackplot(years, EU_emisja.emissions, Chiny_emisja.emissions, Indie_emisja.emissions, USA_emisja.emissions, labels=['EU', 'China', 'India', 'USA'])
ax1.stackplot(years, Swiat_emisja.emissions, alpha = 0.3, labels = ['World'])
ax1.set_ylabel('CO2 emissions (kt)')
ax1.set_xlabel('Year')
ax1.set_title('EU vs China')
plt.legend()
plt.show()
```
